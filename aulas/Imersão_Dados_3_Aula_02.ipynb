```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression='zip')
dados
```

```python
mapa = {'droga': 'composto'}

dados.rename(columns=mapa, inplace=True)

dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[:5]

cod_compostos
```

```python
sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(data=dados.query('composto in @cod_compostos'), x='composto')
ax.set_title('Top 5 Compostos')
plt.show()
```

### Desafio 01: Ordernar o gráfico countplot

```python
cod_compostos
```

```python
sns.set()
plt.figure(figsize=(8, 6))

ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)
ax.set_title('Top 5 compostos', fontdict={'fontsize': 25,
                                          'family': 'serif'})
ax.set_xlabel('Compostos',
              fontdict={'fontsize': 20,
                        'family': 'serif'})
ax.set_xlabel('Compostos',
              fontdict={'fontsize': 20,
                        'family': 'serif'}, labelpad=25)
ax.set_ylabel('Quantidade',
              fontdict={'fontsize': 20,
                        'family': 'serif'}, labelpad=20)
plt.xticks(fontsize=12, family='serif', rotation=10)
plt.yticks(fontsize=12, family='serif')
plt.show()
```

### Desafio 02: Melhorar a visualização alterando o tamanho da fonte...

```python
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)
ax.set_title('Top 5 compostos', fontdict={'fontsize': 25,
                                          'family': 'serif'})
ax.set_xlabel('Compostos',
              fontdict={'fontsize': 20,
                        'family': 'serif'})
ax.set_xlabel('Compostos',
              fontdict={'fontsize': 20,
                        'family': 'serif'}, labelpad=25)
ax.set_ylabel('Quantidade',
              fontdict={'fontsize': 20,
                        'family': 'serif'}, labelpad=20)
plt.xticks(fontsize=12, family='serif', rotation=10)
plt.yticks(fontsize=12, family='serif')
plt.show()
```
